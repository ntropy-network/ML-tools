# Training parallel cross-encoders



```python
configs = TrainConfigs(no_cuda=False, parallelize_cross_encoder=True, model_type='bart', output_dir="outputs", encoder_hidden_layers=2, epochs=6, batch_size=32, dataloader_num_workers=2, negative_labels_per_sample=3, learning_rate=1e-04, early_stopping_epochs=6, validation_data_ratio=0.15, max_length=128)
configs
```
